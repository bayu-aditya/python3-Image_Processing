# **TASK: Import the usual libraries you think you'll need.**

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
def display(gambar):
    fig = plt.figure(figsize = (10,10))
    ax = fig.add_subplot(111)
    ax.imshow(gambar)
    plt.show()
```

```python
plat = cv2.imread('DATA/car_plate.jpg')
plat = cv2.cvtColor(plat, cv2.COLOR_BGR2RGB)
display(plat)
print(plat.shape)
```

# **TASK: Load the haarcascade_russian_plate_number.xml file.**

```python
cascade = cv2.CascadeClassifier('DATA/haarcascades/haarcascade_russian_plate_number.xml')
```

```python
lokasi = cascade.detectMultiScale(plat)
print(lokasi)
```

```python
hasil = plat.copy()

for x,y,w,h in lokasi:
    cv2.rectangle(hasil, (x,y), (x+w, y+h), (255,0,0), 3)
display(hasil)
    
plt.show()
```

## **Set parameter Cascade**

```python
hasil = plat.copy()

lokasi = cascade.detectMultiScale(hasil, scaleFactor = 1.3, minNeighbors = 4)
print(lokasi)
```

```python
for x,y,w,h in lokasi:
    cv2.rectangle(hasil, (x,y), (x+w, y+h), (255,0,0), 3)
display(hasil)
plt.show()
print(hasil.shape)
```
