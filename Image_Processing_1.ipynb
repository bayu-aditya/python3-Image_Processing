# 1. Introduction

```python
import numpy as np
import cv2
import matplotlib.pyplot as plt
```

## 1.1. Convert BGR ke RGB

```python
# Dalam Format BGR
gambar = cv2.imread('DATA/00-puppy.jpg')
plt.imshow(gambar)
plt.show()
```

```python
# Dalam format RGB
gambar = cv2.cvtColor(gambar, cv2.COLOR_BGR2RGB)
plt.imshow(gambar)
plt.show()
```

```python
gambar_gray = cv2.cvtColor(gambar, cv2.COLOR_RGB2GRAY)
plt.imshow(gambar_gray, cmap='gray'); plt.show()
```

```python
print(gambar.shape,'\n',gambar_gray.shape)
```

## 1.2. Resize Image

```python
plt.imshow(cv2.resize(gambar,(600,200)))
plt.show()
```

```python
plt.imshow(cv2.resize(gambar,(0,0),gambar,0.4,0.5))
```

## 1.3. Flipping Image

```python
fig = plt.figure(figsize = (15,4))
fig.suptitle('Flipping Image')
ax1 = fig.add_subplot(131)
ax1.imshow(cv2.flip(gambar, flipCode = 1))

ax2 = fig.add_subplot(132)
ax2.imshow(cv2.flip(gambar, flipCode = 0))

ax2 = fig.add_subplot(133)
ax2.imshow(cv2.flip(gambar, flipCode = -1))
```

# 2. Drawing on Images with OpenCV 
-----

```python
blank = np.zeros((512,512,3), dtype = np.int)
plt.imshow(blank); plt.show()
```

## 2.1. Rectangle

```python
cv2.rectangle(blank, pt1 = (20,20), pt2 = (150,150), 
              color = (0,255,0), thickness = 2)
plt.imshow(blank); plt.show()
```

## 2.2. Cicle

```python
cv2.circle(blank, center = (300,300), radius = 30, 
           color = (0,0,255), thickness = 3)
plt.imshow(blank); plt.show()
```

## 2.3. Line

```python
cv2.line(blank, pt1 = (100,200), pt2 = (200,400),
         color = (255,0,0), thickness = 4)
plt.imshow(blank); plt.show()
```

## 2.4. putText

```python
cv2.putText(blank, text = 'Ini Text', org = (200,100), fontFace = cv2.FONT_HERSHEY_PLAIN,
            fontScale = 5, color = (255,255,0), thickness = 3)
plt.imshow(blank); plt.show()
```

## 2.5. polyLines

```python
point = np.array([[300,400],[400,300],[500,300],[400,400]])
print(point.shape ,'\n', point)
point = point.reshape((-1,1,2))
print(point.shape ,'\n', point)
```

```python
cv2.polylines(blank, pts = [point],
              isClosed = True, color = (0,255,255), thickness = 3)
plt.imshow(blank); plt.show()
```

# 3. Drawing on Images with Mouse
-----

```python
def fungsi_gambar_lingkaran(event,x,y,flags,param):
    if event == cv2.EVENT_LBUTTONDOWN:
        cv2.circle(blank, center = (x,y), radius = 50, color = (250,0,0), thickness = -1)

blank = np.zeros((512,512,3), dtype = np.uint8)

cv2.namedWindow('testing')
cv2.setMouseCallback('testing', fungsi_gambar_lingkaran)

while True:
    cv2.imshow('testing', blank)
    if cv2.waitKey(20) & 0xFF == 27:
        break

cv2.destroyAllWindows()
```

```python
def fungsi_gambar2(event, x, y, flags, param):
    if event == cv2.EVENT_LBUTTONDOWN:
        cv2.circle(blank, center = (x,y), radius = 50, color = (255,255,0), thickness = -1)
    elif event == cv2.EVENT_RBUTTONDOWN:
        cv2.circle(blank, center = (x,y), radius = 50, color = (0,0,255), thickness = -1)
        
blank = np.zeros((512,512,3), dtype = np.uint8)

cv2.namedWindow('nama window')
cv2.setMouseCallback('nama window', fungsi_gambar2)

while True:
    cv2.imshow('nama window', blank)
    if cv2.waitKey(20) & 0xFF == 27:
        break
        
cv2.destroyAllWindows()
```
