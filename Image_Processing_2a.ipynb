**TASK: Open and display the giraffes.jpg image that is located in the DATA folder.**

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
gambar = cv2.imread("DATA/giraffes.jpg")
gambar = cv2.cvtColor(gambar, cv2.COLOR_BGR2RGB)
plt.title('Gambar Original')
plt.imshow(gambar); plt.show()
```

```python
print(gambar.shape)
```

**TASK:Apply a binary threshold onto the image.**

```python
gambar_gray = cv2.cvtColor(gambar, cv2.COLOR_RGB2GRAY)

ret, thres1 = cv2.threshold(gambar_gray, 100, 255, cv2.THRESH_BINARY)

print(ret, '\n', thres1.shape)
```

```python
plt.imshow(thres1, cmap = 'gray'); plt.show()
```

**TASK: Create a low pass filter with a 4 by 4 Kernel filled with values of 1/10 (0.01) and then use 2-D Convolution to blur the giraffer image (displayed in normal RGB)**

```python
kernel = np.full((4,4), 0.1)
print(kernel)
```

```python
gambar_blur = cv2.filter2D(gambar, -1, kernel)
plt.imshow(gambar_blur)
```

**TASK: Create a Horizontal Sobel Filter (sobelx from our lecture) with a kernel size of 5 to the grayscale version of the giaraffes image and then display the resulting gradient filtered version of the image.**

```python
sobelx = cv2.Sobel(gambar_gray, cv2.CV_64F, 1, 0, ksize = 5)
sobely = cv2.Sobel(gambar_gray, cv2.CV_64F, 0, 1, ksize = 5)

fig = plt.figure(figsize = (20,20))
ax1 = fig.add_subplot(121)
ax2 = fig.add_subplot(122)

ax1.imshow(sobelx, cmap = 'gray')
ax2.imshow(sobely, cmap = 'gray'); plt.show()
```

**TASK: Plot the color histograms for the RED, BLUE, and GREEN channel of the giaraffe image. Pay careful attention to the ordering of the channels.**

```python
gambar = cv2.imread('DATA/giraffes.jpg')
warna = ('b','g','r')

for no, color in enumerate(warna):
    hist = cv2.calcHist([gambar], channels = [no], mask = None, 
                        histSize = [256], ranges = [0,256])
    plt.plot(hist, c = color, label = color)
    plt.xlim([0,256])

plt.legend()
plt.show()
```
